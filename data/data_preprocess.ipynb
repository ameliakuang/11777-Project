```python
class LanguageComputationalSequence(object):
    def __init__(self, key, features, intervals):
        self.key = key
        self.features = features
        self.intervals = intervals
```

```python
import os             
all_txt_files = os.listdir("/content/drive/MyDrive/moseas_french/moseas_spanish/B1_2_Transcript_timestamped_sentences")
% cd "/content/drive/MyDrive/moseas_french/moseas_spanish/B1_2_Transcript_timestamped_sentences"
```

```python
def process_line(i, line):
    elems = line.strip().split('___')
    print(elems)
    key = '{}_{}'.format(elems[0], i)
    print(elems[1], float(elems[1]))
    intervals = (float(elems[1]), float(elems[2]))
    sentence = elems[3]
    return (key, intervals, sentence)
```

```python
highlevel = []

for txt_file in all_txt_files:
    with open(txt_file) as f:
        lines = f.readlines()
        f.close()
        for i, line in enumerate(lines):
            key, intervals, sentence = process_line(i, line)
            # print(txt_file, key, intervals, sentence)
            key_maps = []
            features = sentence.split(' ')
            compseq = LanguageComputationalSequence(key, features, intervals)
            highlevel.append(compseq)
```

```python
for d in highlevel:
    print(d.key, d.features, d.intervals)
```

```python
word2id = {}
```

```python
for d in highlevel:
    for word in d.features:
        if word not in word2id:
            word2id[word] = len(word2id)
```

```python
from collections import defaultdict

emotions = ['anger', 'confident', 'disgust', 'dominant', 'eloquent', 'entertaining', 'fear', 'gender', 'happiness', 'humorous', 'is_opinion', 'narcissist', 'nervous', 'passionate', 'persuasive', 'relaxed', 'reserved', 'sadness', 'sarcastic', 'surprise']
emotion_map = {}
for i, emotion in enumerate(emotions):
    emotion_map[emotion] = i
```

```python
emotion_map
```

```python
import csv
import re

result_dict = {}
emo_map = {}
for emo in emotions:
    emo_map[emo] = 0
paths = ["/content/drive/MyDrive/moseas_french/moseas_spanish/Batch_3851530_batch_results.csv", "/content/drive/MyDrive/moseas_french/moseas_spanish/Batch_3860330_batch_results.csv"]
for path in paths:
  with open(path) as f:
      reader = csv.DictReader(f)
      for row in reader:
          video_id = row['Input.VIDEO_ID'].split('/')[-1]
          rgx_res = re.findall('[0-9]:[0-9][0-9]:[0-9][0-9]', video_id)
          idx = video_id.index(rgx_res[0])
          key = video_id[:idx-1]
          if key not in result_dict:
              result_dict[key] = emo_map.copy()
          for emo in emotions:
              value = row['Answer.{}'.format(emo)]
              if value in {'1', 'checked', 'yes'}:
                  result_dict[key][emo] += 1
          for i in range(1, 4):
              result_dict[key]['secret_word{}'.format(i)] = row['Answer.secret_word{}'.format(i)]
          if row['Answer.sentiment'].lstrip('-').isnumeric():
              if 'sentiment' not in result_dict[key]:
                  result_dict[key]['sentiment'] = [int(row['Answer.sentiment'])]
              else:
                  result_dict[key]['sentiment'].append(int(row['Answer.sentiment']))
result_dict
```

```python
from collections import Counter
import numpy as np
for key, val in result_dict.items():
    for emo in emotions:
        if val[emo] > 1:
            val[emo] = 1
        else:
            val[emo] = 0
    result_dict[key] = val
    sentiments = result_dict[key]['sentiment']
    # print(sentiments, np.average(sentiments))
    # count = Counter(sentiments)
    # result_dict[key]['sentiment'] = count.most_common(1)[0][0]
    result_dict[key]['sentiment'] = np.average(sentiments)
```
